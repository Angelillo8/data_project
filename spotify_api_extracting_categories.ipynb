```python
def fetch_playlists(categorias):
    """
    Returns the user's playlists.
    """
        
    id = []
    name = []
    num_tracks = []
    category = []
    
    # Make the API request
    for i in categorias:
        playlist_results = sp.category_playlists(category_id=i, limit=50)
        for k in playlist_results["playlists"]['items']:
            id.append(k['id'])
            name.append(k['name'])
            num_tracks.append(k['tracks']['total'])
            category.append(i)


    # Create the final df   
    df_playlists = pd.DataFrame({"id":id, "name": name, "#tracks": num_tracks, "playlits_category":category})
    return df_playlists

```

```python
def fetch_playlist_tracks(playlist,df):
    """
    Returns the tracks for the given playlist.
    """
        
    offset = 0
    tracks = []
    
    # Make the API request
    while True:
        content = sp.playlist_tracks(playlist[0], fields=None, limit=100, offset=offset, market=None)
        tracks += content['items']

        if content['next'] is not None:
            offset += 100
        else:
            break
    
    track_id = []
    track_name = []
    track_popularity = []
    track_duration = []
    track_artist = []
    track_album = []
    track_release_date = []
    track_playlist = []
    track_playlist_id = []
    track_playlist_category = []
    
    for track in tracks:
        track_id.append(track['track']['id'])
        track_name.append(track['track']['name'])
        track_popularity.append(track['track']['popularity'])
        track_duration.append(track['track']['duration_ms'])
        track_artist.append(track['track']['artists'][0]["name"])
        track_album.append(track['track']['album']["name"])
        track_release_date.append(track['track']['album']["release_date"])
        track_playlist.append(playlist[1])
        track_playlist_id.append(playlist[0])
        track_playlist_category.append(playlist[3])
    # Create the final df
    df_playlists_tracks = pd.DataFrame({"id":track_id, "name": track_name, 
                                        "popularity": track_popularity, "duration": track_duration,
                                       "artist":track_artist, "album":track_album,
                                       "release_date": track_release_date, "playlist":track_playlist,
                                       "playlist_id":track_playlist_id, "playlist_category":track_playlist_category})
    return df.append(df_playlists_tracks,ignore_index=True)


```

```python
def fetch_audio_features(alltracks):
    """
    Returns the selected audio features of every track, 
    for the given playlist.
    """
    index = 0
    audio_features = []
    
    # Make the API request
    while index < alltracks.shape[0]:
        audio_features += sp.audio_features(alltracks.iloc[index:index + 50, 0])
        index += 50
    
    # Append the audio features in a list
    features_list = []
    for features in audio_features:
        features_list.append([features['danceability'],features['energy'], 
                              features['key'], features['loudness'], 
                              features['mode'],features['speechiness'],
                             features['acousticness'], features['instrumentalness'],
                             features['liveness'], features['valence'],
                             features['tempo'], features['time_signature']])
    
    df_audio_features = pd.DataFrame(features_list, columns=['danceability', 'energy', "key",
                                                             "loudness", "mode", "speechiness",
                                                            "acousticness", "instrumentalness",
                                                            "liveness", "valence", "tempo", "time_signature"])
    
    # Create the final df, using the 'track_id' as index for future reference
    df_playlist_audio_features = pd.concat([alltracks, df_audio_features], axis=1)
    df_playlist_audio_features.set_index('track_id', inplace=True, drop=True)
    
    return df_playlist_audio_features

```
