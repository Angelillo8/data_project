```python
def scrap_stuff (dates,df):
    url = "https://spotifycharts.com/regional/global/weekly/"
    
    weeks = []
    for i in dates:
        link = url + i
        
        html = requests.get(link)
        html_soup = BeautifulSoup(html.content,"html.parser")

        track_art = html_soup.find_all("td",class_="chart-table-track")
        track_art = [i.get_text().split("\n") for i in track_art]
        track_art = pd.DataFrame(track_art)
        track_art = track_art.drop([0,3],axis=1)
        id = []
        for i in range(6,205):
            id.append(html_soup.select("a")[i].attrs["href"])
        songdata = pd.DataFrame({"week":i,"id":id})
        data = pd.concat([songdata, track_art], axis=1)
        df.append(data,ignore_index=True)
    return df
```

```python
url = "https://spotifycharts.com/regional/global/weekly/"
id = []
artist=[]
tracks=[]
date = []
for i in fechas:
    link = url + i
        
    html = requests.get(link)
    html_soup = BeautifulSoup(html.content,"html.parser")

    track_art = html_soup.find_all("td",class_="chart-table-track")
    track_art = [i.get_text().split("\n") for i in track_art]
    for j in range(len(track_art)):
        tracks.append(track_art[j][1])
        artist.append(track_art[j][2])
        date.append(i)
    for i in range(6,206):
        id.append(html_soup.select("a")[i].attrs["href"])
datos = pd.DataFrame({"date":date,"id":id})
print(datos)
```

```python
songdata = pd.DataFrame({"week":date,"id":id,"track":tracks,"artist":artist})
top.append(songdata,ignore_index=True)
```

```python
top200 = scrap_stuff(fechas,top)
```
